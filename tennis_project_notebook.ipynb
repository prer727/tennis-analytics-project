# Tennis Game Analytics Starter Notebook

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
```

```
categories = pd.read_csv("categories.csv")
competitions = pd.read_csv("competitions.csv")
complexes = pd.read_csv("complexes.csv")
venues = pd.read_csv("venues.csv")
competitors = pd.read_csv("competitors.csv")
rankings = pd.read_csv("competitor_rankings.csv")

print(categories.shape, competitions.shape, complexes.shape, venues.shape, competitors.shape, rankings.shape)
```

```
comp_rank = competitors.merge(rankings, on="competitor_id", how="inner")
comp_rank.head()
```

```
top5 = comp_rank.sort_values("rank").head(5)
top5
```

```
plt.figure(figsize=(8,4))
sns.barplot(data=top5, x="name", y="points")
plt.xticks(rotation=45, ha="right")
plt.title("Top 5 Ranked Competitors")
plt.show()
```

**Why this chart?**  
Bar charts make leaderboard comparison easy.

**Insight:**  
The top-ranked competitors also lead on points.

**Business impact:**  
Useful for quick ranking dashboards and leaderboard reporting.


```
country_counts = competitors["country"].value_counts().reset_index()
country_counts.columns = ["country", "count"]
country_counts
```

```
plt.figure(figsize=(8,4))
sns.barplot(data=country_counts, x="country", y="count")
plt.xticks(rotation=45, ha="right")
plt.title("Competitors per Country")
plt.show()
```

**Why this chart?**  
It compares country participation.

**Insight:**  
Some countries contribute more teams than others.

**Business impact:**  
Helps country-wise analysis and federation-level summaries.
